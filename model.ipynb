# 2. Reading the data
---

One of the problems with the dataset is that each vital signal was sampled with a different frequency. This way, we need to find a way to put all of them in the same dimensions.

A first solution is to artificially resample all signals using the Fourier method. To this end, we need to make the hypothesis that the signals are periodic.

```python
def resampleSignal(arr, N):
    """
        Resampling signal (time domain). Using the scipy Fourier method to resample signals
        to the same number of points. Because a Fourier method is used, the signal is
        assumed to be periodic.

        arr = numpy array with data in columns
        N  = number of points after resampling 
    """

    arr_new = np.zeros((N, arr.shape[1]))

    for col in range(arr.shape[1]):
        arr_new[:, col] = signal.resample(arr[:, col], N).astype(np.float64)
        #arr_new[:, col] = np.clip(arr_new[:, col], 0, 1)
    
    return arr_new
```

Here is the helper function to read the signal, resample and normalize it.

```python
def readSignal(filepath, N=None, normalize=False):

    if 'ACC.csv' in filepath:
        df = pd.read_csv(filepath, header=None, names=['x', 'y', 'z'])
    else:
        df = pd.read_csv(filepath, header=None, names=[filepath.split('/')[-1].split('.')[0]])

    arr = df.values[2:, :]

    if N is not None:
        arr = resampleSignal(arr, N)

    if normalize:
        scaler = MinMaxScaler((0,1))
        arr = scaler.fit_transform(arr)

    return arr
```

- For the final exam: 3h duration -> 180 * 60 * 5 = 54000 points (frequency of 5Hz)

Now let's create the helper function to read and normalize the grades (our labels):

```python
def readGrades(filepath):

    df = pd.read_csv(filepath)

    st = df['Student']
    grades = df.drop(['Student'], axis=1)

    scaler = MinMaxScaler()
    grades = scaler.fit_transform(grades)
    grades = pd.DataFrame(grades, columns=['Midterm 1', 'Midterm 2', 'Final'])
    
    df = pd.concat([st, grades], axis=1)

    return df


filepath = 'dataset/grades.csv'

df = pd.read_csv(filepath)
seq_len = len(df)
data = df.values

fig, axs = plt.subplots(1,2)

col = 'Final'

# original
axs[0].plot(np.arange(seq_len), df[col])
axs[0].set_title('Original')

# normalized

df_new = readGrades(filepath)
axs[1].plot(np.arange(seq_len), df_new[col])
axs[1].set_title('Normalized')

plt.tight_layout
plt.show()
```

# 3. Building the dataset
---

Once we can read the data from each .csv file individually, we need to put everything together. We create a custom dataset using the _Dataset_ module of PyTorch, that will help the training and evaluation in the future.

```python
class myDataset(Dataset):

    def __init__(self, N=54000):
        super().__init__()
        self.N = N
        self.X, self.Y = self.initDataset()

    def __len__(self):
        assert len(self.X) == len(self.Y)
        return len(self.X)
    
    def __getitem__(self, idx):
        x, y = self.X[idx], self.Y[idx]
        return th.from_numpy(x).float(), th.from_numpy(y).float()
    
    def initDataset(self):
        student = ['S'+str(i) for i in range(1, 11)]
        exam = ['Midterm 1', 'Midterm 2', 'Final']
        signals = ['ACC', 'BVP', 'EDA', 'HR', 'IBI', 'TEMP']

        X, Y = [], []
        
        grades = readGrades('dataset/grades.csv')

        for stu in tqdm(student, desc='Loading data'):

            for exa in exam:

                aux = np.zeros((self.N, 1))

                for sig in signals:
                    
                    filepath = 'dataset/{}/{}/{}.csv'.format(stu, exa, sig)
                    aux = np.hstack((aux, readSignal(filepath, self.N, normalize=True)))


                X.append(aux[:, 1:])
                Y.append(grades[grades['Student'] == stu][exa].values)

        return X, Y

dataset = myDataset()
```

Once the dataset is loaded, let's check its shapes and ranges.

```python
# Testing dataset
train_loader = DataLoader(dataset, batch_size=5, shuffle=True)

# X -> (batch_size, seq_len, num_features)
# Y -> (batch_size, 1)

X, Y = next(iter(train_loader))

print('X dim: {} | Y dim: {}'.format(X.size(), Y.size()))
print('X min, max: {}, {} | Y min, max: {}, {}'.format(X.min(), X.max(), Y.min(), Y.max()))
```

# 4. Building the model
---

Before diving into the model itself, let's understand the RNN module and how it modifies the input.

```python
# RNN module
batch_size = X.shape[0]
seq_len = X.shape[1]
input_size = X.shape[2]

hidden_size = 3
num_layers = 1

rnn = th.nn.RNN(input_size, hidden_size, num_layers, batch_first=True)

h0 = th.zeros((num_layers, batch_size, hidden_size)) # init hidden layer
h0 = h0.float()

out, hn = rnn(X[:,0,:], h0[:,0,:])
print('out dim: {} \n hn dim: {}\n'.format(out.size(), hn.size()))

print('Initial state (h0): {}\n   Last state (hn): {}\n'.format(h0[:,0,:].flatten(), hn.flatten()))
```

Ok, let's try to build a first version of our model. To start basic, we are going just to create some layers to reduce the _seq\_len_ dimension (of 54000), pass through an RNN layer, take the last predicted time step and using it to make a prediction of the grade using a sigmoid function:

```python
class GradePredictor(nn.Module):
    
    def __init__(self, input_size, hidden_size, n_layers):
        super(GradePredictor, self).__init__()
        
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.n_layers = n_layers

        self.conv1 = nn.Conv1d(in_channels=self.input_size, out_channels=self.input_size, kernel_size=2, stride=2)
        self.conv2 = nn.Conv1d(in_channels=self.input_size, out_channels=self.input_size, kernel_size=4, stride=4)
        self.conv3 = nn.Conv1d(in_channels=self.input_size, out_channels=self.input_size, kernel_size=8, stride=8)

        self.rnn = nn.RNN(input_size=self.input_size, hidden_size=self.hidden_size, num_layers=self.n_layers, batch_first=True)

        self.fc = nn.Linear(self.hidden_size, 1)
        self.sigmoid = nn.Sigmoid()

    def init_hidden(self, batch_size):
        self.h = th.zeros(self.n_layers, batch_size, self.hidden_size).to(device)
    
    def forward(self, x, h0=None):
        
        out = x.permute(0, 2, 1) # out -> (batch_size, input_size, seq_len)

        out = self.conv1(out)
        out = self.conv2(out)
        out = self.conv3(out)

        out = out.permute(0, 2, 1) # out -> (batch_size, seq_len, input_size)

        if h0 is not None:
            self.h = h0
        else:
            self.init_hidden(out.shape[0])

        out, self.h = self.rnn(out, self.h) # out -> (batch_size, seq_len, hidden_size)
                                            # hn  -> (num_layers, batch_size, hidden_size)

        out = out[:,-1,:] # out -> (batch_size, 1, hidden_size)
        out = self.fc(out)
        out = self.sigmoid(out)

        return out, self.h
```

Testing the dimensions:

```python
model = GradePredictor(input_size=input_size, hidden_size=hidden_size, n_layers=num_layers).to(device)

X = X.to(device)
Y = Y.to(device)

out, hn = model(X)
print('out dim: {} | hn dim: {} | Y dim: {}\n'.format(out.size(), hn.size(), Y.size()))
print(th.concat((out, Y), axis=1))
```

We are going to use the _Mean Squared Error_ as a loss function, once we want to predict a real number at the output (the student's grade).

```python
criterion = nn.MSELoss() # mean squared error
loss = criterion(out, Y)

print('\nLoss: {}'.format(loss.item()))
```

# 5. Trainer and Evaluator
---

These are the helper functions to train and evaluate our model. Nothig special here.

```python
def trainer(model, device, dataloader, criterion, optimizer, threshold=0.02):

    train_loss, train_acc = 0.0, 0.0

    model.train()
    
    for x, y in dataloader:

        x, y = x.to(device), y.to(device)
        
        output, _ = model(x)
        loss = criterion(output, y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        train_loss += loss.item()

        train_acc += ((th.abs(output - y) < threshold).sum().item()) / y.size(0)

    return train_loss/len(dataloader), train_acc/len(dataloader)
```

```python
def evaluator(model, device, dataloader, criterion, threshold=0.02):

    valid_loss, val_acc = 0.0, 0.0

    model.eval()

    for x, y in dataloader:

        x, y = x.to(device), y.to(device)

        output, _ = model(x)
        loss = criterion(output, y)

        valid_loss += loss.item()
        
        val_acc += ((th.abs(output - y) < threshold).sum().item()) / y.size(0)

    return valid_loss/len(dataloader), val_acc/len(dataloader)
```

# 6. The K-Fold method
---

Because we haven't a lot of data available to split into _train_ and _test_ datasets, we are going to implement a _k-fold_ method to take advantage of everything for both learning and evaluation.

```python
# K-fold method

th.manual_seed(42)

splits = KFold(n_splits=5, shuffle=True, random_state=42) # Creating folds

n_epoch = 2

for epoch in range(n_epoch):

    for fold, (train_idx, test_index) in enumerate(splits.split(dataset.X)):

        train_sampler = SubsetRandomSampler(train_idx)
        test_sampler = SubsetRandomSampler(test_index)

        print('Epoch: {} | Fold {}: | train_sampler idx: {} | test_sampler idx: {}'.format(epoch, fold, train_sampler.indices, test_sampler.indices))

    print('')
```

```python
# K-fold method

th.manual_seed(42)

splits = KFold(n_splits=5, shuffle=True, random_state=42) # Creating folds

n_epoch = 2

for epoch in range(n_epoch):

    for fold, (train_idx, test_index) in enumerate(splits.split(dataset.X)):

        print('Epoch {}, Fold {}'.format(epoch, fold))

        train_sampler = SubsetRandomSampler(train_idx)
        test_sampler = SubsetRandomSampler(test_index)

        train_loader = DataLoader(dataset, batch_size=6, sampler=train_sampler)
        for X,Y in train_loader:
            print('Train: Xshape ({}), Yshape ({})'.format(X.size(), Y.size()))

        test_loader = DataLoader(dataset, batch_size=6, sampler=test_sampler)
        for X,Y in test_loader:
            print('Test: Xshape ({}), Yshape ({}) \n'.format(X.size(), Y.size()))

    print('')
```

# 7. First try
---

Now we can train and evaluate our model:

```python
# K-fold method

th.manual_seed(42)

splits = KFold(n_splits=5, shuffle=True, random_state=42) # Creating folds

n_epoch = 150

# Model
input_size = 8
hidden_size = 10
num_layers = 2

model = GradePredictor(input_size=input_size, hidden_size=hidden_size, n_layers=num_layers)
model = model.to(device)

# Loss function and optimizer
criterion = nn.MSELoss() # mean squared error
optimizer = th.optim.Adam(model.parameters(), lr=1e-3)

# Model info
trainInfo = {'epoch': [], 'fold': [], 'loss': [], 'acc': []}
evalInfo = {'epoch': [], 'fold': [], 'loss': [], 'acc': []}

for epoch in range(n_epoch):

    global_loss, global_acc = 0.0, 0.0

    for fold, (train_idx, test_index) in enumerate(splits.split(dataset.X)):

        train_sampler = SubsetRandomSampler(train_idx) # Index of each fold
        test_sampler = SubsetRandomSampler(test_index)

        train_loader = DataLoader(dataset, batch_size=6, sampler=train_sampler) # Creating trainloaders with folder's indexes
        test_loader = DataLoader(dataset, batch_size=6, sampler=test_sampler)

        train_loss, train_acc = trainer(model, device, train_loader, criterion, optimizer, threshold=0.02)
        global_loss += train_loss
        global_acc += train_acc

        eval_loss, eval_acc = evaluator(model, device, test_loader, criterion, threshold=0.02)

        trainInfo['epoch'].append(epoch), trainInfo['fold'].append(fold), trainInfo['loss'].append(train_loss), trainInfo['acc'].append(train_acc)
        evalInfo['epoch'].append(epoch), evalInfo['fold'].append(fold), evalInfo['loss'].append(eval_loss), evalInfo['acc'].append(eval_acc)

    if epoch % 10 == 0:
        print('Epoch {}/{}: loss {:.4f} | acc {:.4f}'.format(epoch, n_epoch, global_loss/6, global_acc/6))
```

It took less than 1 minute to train using the GPU (~6 minutes on CPU)! Let's take a look in the learning curves:

```python
def learningCurves(infoDict):
    df = pd.DataFrame(infoDict)
    df.set_index('epoch', inplace=True)

    epoch_loss, epoch_acc = [], []

    for epoch in range(n_epoch):
        loss = df.loc[epoch]['loss'].sum()/len(df.loc[epoch])
        acc = df.loc[epoch]['acc'].sum()/len(df.loc[epoch])

        epoch_loss.append(loss), epoch_acc.append(acc)

    return np.array(epoch_loss), np.array(epoch_acc)
```

```python
train_loss, train_acc = learningCurves(trainInfo)
eval_loss, eval_acc = learningCurves(evalInfo)

print('Train acc: {:.2f}% -> Eval acc: {:.2f}%'.format(train_acc[-1]*100, eval_acc[-1]*100))

fig, axs = plt.subplots()

axs.plot(train_loss, label='Train')
axs.plot(eval_loss, label='Eval')
axs.set_title('Loss across epochs')
axs.set_ylabel('Loss')
axs.set_xlabel('Epoch')
axs.legend()

plt.tight_layout(pad=.2)
plt.show()
```
